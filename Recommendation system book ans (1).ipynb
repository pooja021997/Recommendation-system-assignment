# Recommendation system-book.csv

```python
import pandas as pd
import numpy as np
```

```python
book=pd.read_csv(r'C:\Users\user\Downloads\book.csv',encoding='ISO-8859-1')
book
```

```python
book.drop(['Unnamed: 0'],axis='columns',inplace=True)
book
```

```python
book_df=book.rename(columns={'User.ID':'userid','Book.Title':'Book Title','Book.Rating':'Book Rating'},inplace=False)
book_df
```

```python
#unique users in the dataset
book_df['userid'].unique()
```

```python
#number of unique users in the data set
len(book_df['userid'].unique())
```

```python
#unique books in the data set
book_df['Book Title'].unique()
```

```python
#number of unique book in the data set
len(book_df['Book Title'].unique())
```

```python
user_book_df=book_df.pivot_table(index='userid',
                                 columns='Book Title',
                                 values='Book Rating').reset_index(drop=False)
user_book_df
```

```python
user_book_df.index=book_df['userid'].unique()
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0,inplace=True)
user_book_df
```

```python
#calculating Cosine Similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine , correlation
```

```python
user_sim=1-pairwise_distances(user_book_df.values,metric='cosine')
user_sim
```

```python
#Store the result in the data frame
user_sim_df=pd.DataFrame(user_sim)
```

```python
#Set he index and column names to user ids
user_sim_df.index=book_df.userid.unique()
user_sim_df.columns=book_df.userid.unique()
```
